# MoodScriber Data Challenge

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Let's start by reading in the Input.xlsx file into a pandas dataframe**

```python
df = pd.read_excel("Input.xlsx")
```

```python
# Diplaying first 10 rows of the dataframe
df.head(10)
```

## Overall Mood Distribution


```python
# Mood Trends
fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(x=df['Mood'], ax=ax)
plt.title('Mood Distribution')
plt.xlabel('Mood')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

# Distribution of moods by day

```python
# Distribution of moods by day
plt.figure(figsize=(10, 6))
sns.set_palette("husl")  
sns.set_style("whitegrid")  

# Define the order of days of the week for sorting
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

sns.countplot(data=df, x='Day', hue='Mood', order=order)
plt.title('Distribution of Moods by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.xticks(rotation=45) 
plt.legend(title='Mood', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Sleep Quality Vs Mood

## Mood Correlation - Activity 1

```python
mood_activity_count = df.groupby(['Mood', 'Activity-1']).size().unstack(fill_value=0)

plt.figure(figsize=(8,6))
sns.heatmap(mood_activity_count, cmap='coolwarm', annot=True, fmt='d')
plt.title('Activity-1 and Mood Correlation')
plt.xlabel('Activity-1')
plt.ylabel('Mood')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Mood', hue='Activity-1', palette='Set2')
plt.title('Activity-1 and Mood Correlation')
plt.xlabel('Mood')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Activity-1', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
